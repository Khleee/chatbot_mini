```python
import pandas as pd
import json
```

```python
with open('data/final2_인텐트 목록(187개).json', 'r',encoding="utf-8") as f:
    json_data = json.load(f)
```

```python
# intent_list 생성
B = pd.read_csv("data/final2_title_node(187개).csv")
intent_list = list(B["title"])

# intent_dic 생성
intent_dic = {}
for i,x in enumerate(intent_list):
    intent_dic[x] = i

print(len(intent_dic))
```

```python
intent_dic
```

```python
ALL = pd.DataFrame(columns = ["example","label"])
```

```python
# 미리 df 만들기
A = pd.DataFrame(columns = ["example","label"])
B = pd.DataFrame(columns = ["example","label"])
C = pd.DataFrame(columns = ["example","label"])
```

```python
for x in json_data: # json data 키 하나씩
    for y in json_data[x]: # x에 해당하는 유사문장들 하나씩
        ALL.loc[len(ALL)] = [y,x]
```

```python
ALL
```

```python
# shuffle 후 index 초기화
ALL = ALL.sample(frac=1).reset_index(drop=True)
```

```python
ALL
```

```python
example = list(ALL["example"])
label = list(ALL["label"])
```

```python
### 라벨마다 5개씩 test 데이터셋에 집어넣자
for w in intent_list: # [인텐트 명 리스트]
    temp_num = 0

    for x,z in zip(example,label): # ([샬라샬라,...],[샬라샬라,...],[14,15...]) <- token_num이 순서대로 안있고 섞여있음
        if w == z and temp_num <= 5:
            # 251*5 -> 1255개
            C.loc[len(C)] = [x,z]
            temp_num += 1
        elif w == z and temp_num%6 == 0:
            B.loc[len(B)] = [x,z]
            temp_num += 1
        elif w == z and temp_num%6 != 0:
            A.loc[len(A)] = [x,z]
            temp_num += 1
```

```python
# 개수 확인
print(len(A),len(B),len(C))
```

```python
# 저장

# train
A.to_csv("data/final2_train2.csv",encoding="utf-8-sig",index = False)

# valid
B.to_csv("data/final2_valid2.csv",encoding="utf-8-sig",index = False)

# test
C.to_csv("data/final2_test2.csv",encoding="utf-8-sig",index = False)
```
